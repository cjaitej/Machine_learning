```python
import pandas as pd

data = pd.read_csv("/content/sample_data/california_housing_train.csv")

data.head()
```

# New Section

```python
import matplotlib.pyplot as plt

for i in data.columns:
  plt.xlabel(i)
  plt.ylabel("median_house_value")
  plt.scatter(data[i],data["median_house_value"])
  plt.show()
```

```python
import numpy as np 

train_x = np.array(data["median_income"])

train_y = np.array(data["median_house_value"])
```

```python
train_x = train_x.reshape(17000,-1)

train_x.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(normalize = True)

model.fit(train_x, train_y)
```

```python
test_data = pd.read_csv("/content/sample_data/california_housing_test.csv")

test_x = np.array(test_data["median_income"]).reshape(3000, -1)
test_y = np.array(test_data["median_house_value"])
```

```python
model.score(test_x, test_y)
```

```python
y_pred = model.predict(test_x)

plt.scatter(test_data["median_income"], test_data["median_house_value"], color = "green")
plt.plot(test_x, y_pred, color = "red")

plt.show()
```
